```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sea
```

```python
df = pd.read_csv('datasets/price_by_state.csv', sep=';', index_col=0)
```

```python
df.head()
```

```python
#df['Preço Médio'] = df['Preço Médio'].apply(lambda x: str(x))
df['Preço Médio'].fillna(0, inplace=True)
```

```python
fuel = df.pivot("Estado - Sigla", "Produto", "Preço Médio")
fig, ax = plt.subplots(figsize=(18,10))
sea.heatmap(fuel, linewidths=1, linecolor='gray', cmap="YlGnBu", annot=True, ax=ax, fmt="f")
ax.set_title("Heatmap - Preço médio do combustível por tipo e estado (Em R$)",fontdict={'size':14})
ax.xaxis.set_label_text("Tipo de combustível",fontdict= {'size':14})
ax.yaxis.set_label_text("Estado",fontdict= {'size':14})
```
