```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df = pd.read_csv('datasets/price_by_state.csv', sep=';', index_col=0)
```

```python
df.head(15)
```

```python
df['Preço Médio'] = df['Preço Médio'].apply(lambda x: round(x, 2))
```

```python
#df['Preço Médio'] = df['Preço Médio'].apply(lambda x: str(x))
df['Preço Médio'].fillna(0, inplace=True)
```

```python
fuel = df.pivot("Estado - Sigla", "Produto", "Preço Médio")
fig, ax = plt.subplots(figsize=(15,15))
sea.heatmap(fuel, linewidths=0.25, linecolor='black', cmap="mako_r", annot=True, annot_kws={"fontsize":14}, ax=ax, fmt="f")
title = plt.figtext(.5,.9,"Preço médio do combustível por tipo e estado (Em R$) - 1° Sem 2019", fontsize=14, ha='right')
title.set_position([.5, 0.92])
ax.xaxis.set_label_text("Tipo de combustível",fontdict= {'size':16})
ax.yaxis.set_label_text("Estado",fontdict= {'size':16})
```

```python
fig.savefig('images/fuel2019.jpg')
```
